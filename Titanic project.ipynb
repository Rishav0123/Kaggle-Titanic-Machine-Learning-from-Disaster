```python
mean_age=round(df.groupby('Pclass').mean()['Age'])
```

```python
mean_age
```

```python
df.head()
```

```python
def impute_age(cols):
   Age = cols[0]
   Pclass = cols[1]
   if pd.isnull(Age):
       if Pclass == 1:
           return mean_age[Pclass]
       elif Pclass == 2:
           return mean_age[Pclass]
       else:
           return mean_age[Pclass]
   else:
       return Age

df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis = 1)
```
